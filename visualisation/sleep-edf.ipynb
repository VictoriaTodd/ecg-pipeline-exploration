```python
import numpy as np
import mne
import matplotlib.pyplot as plt
```

Import a file from the sleep cassette data and preload it:

```python
raw_cassette = mne.io.read_raw_edf("../../../Data/sleep-edf-database-expanded-1.0.0/sleep-cassette/SC4001E0-PSG.edf", preload=True)
```

Print some info about the channels:

```python
print(raw_cassette.info)
print(raw_cassette.info.ch_names)
```

Plot each channel individually so we can view it:

```python
raw_cassette.plot(picks=['EEG Fpz-Cz'])
raw_cassette.plot(picks=['EEG Pz-Oz'])
raw_cassette.plot(picks=['EOG horizontal'])
raw_cassette.plot(picks=['Resp oro-nasal'])
raw_cassette.plot(picks=['EMG submental'])
raw_cassette.plot(picks=['Temp rectal'])
raw_cassette.plot(picks=['Event marker'])
```

Test plotting with plt:

```python
idx = raw_cassette.ch_names.index("Resp oro-nasal")
event_data, times = raw_cassette[idx, :]

plt.plot(times, event_data[0])
plt.title("Resp oro-nasal channel")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.show()
```

Unclear what the event marker channel is:

```python
idx = raw_cassette.ch_names.index("Event marker")
event_data, times = raw_cassette[idx, :]

import matplotlib.pyplot as plt
plt.plot(times, event_data[0])
plt.title("Event marker channel")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.show()
```

```python
idx = raw_cassette.ch_names.index("Event marker")
event_data, times = raw_cassette[idx, :]

plt.figure(figsize=(12, 3))
plt.plot(times, event_data[0], '.', markersize=2)
plt.title("Marker Channel")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```
